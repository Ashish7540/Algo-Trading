# Returns and Log returns

```python
# Creating sample dataframe
df=pd.DataFrame(index=[2016, 2017, 2018], data=[100, 50, 95], columns=["Price"])
df
```

```python
# Calculate Simple interest and mean of simple interest
simple_returns=df.pct_change().dropna()
simple_returns
```

```python
simple_returns.mean()
```

```python
# Try to find out final value using simple returns
Vf = 100*1.2*1.2
Vf      #Note that final value do not match with data frame final value
# So we need log returns here
```

```python
#Finding lof returns and mean log returns
log_returns=np.log(df/df.shift(1)).dropna()
log_returns
```

```python
log_returns.mean()
```

```python
# Now finding final value using log returns
Vf = 100*np.exp(log_returns.mean()*2)
Vf
# Note here we got correct final value
```

```python
SPY = yf.download("SPY")
SPY
```

```python
spy = SPY.Close.to_frame()
spy
```

```python
spy.plot(figsize=(15,6))
```

```python
# Used to find rooling operations for the window period (like mean, median, mode, max, min....)
spy_roll=spy.rolling(window=10)
spy_roll
```

```python
#Mean of past 10 days
spy_roll_mean = spy_roll.mean()
spy_roll_mean
```

```python
spy_roll_median=spy_roll.median()
spy_roll_median
```

```python
# min_periods attribute finds the rolling value starting from the min period row, but as we pass window period, it will be counting window period only
spy.rolling(window=10, min_periods=5).mean().head(20)
```

```python
spy.rolling(window=10).mean().head(20)
```

```python
spy["SMA50"]=spy.rolling(window=50).mean()
```

```python
spy
```

```python
spy.plot(figsize=(15,8), fontsize=15)
```

```python
# Here we need to specify spy.Close because now we have 2 columns
spy["SMA200"]=spy.Close.rolling(window=200).mean()
```

```python
spy
```

```python
spy.plot(figsize=(15,8), fontsize=15)
```

```python
# plotting Exponential moving avg (EMA100)
spy["EMA100"]=spy.Close.ewm(span=100).mean()
spy.plot(figsize=(15,8), fontsize=15)
```
